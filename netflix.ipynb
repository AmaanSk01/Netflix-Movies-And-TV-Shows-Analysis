```python
import numpy as np
```

```python
# Read the CSV file into a pandas DataFrame
```

```python
df = pd.read_csv("netflix_data2.csv")
```

```python
# Print the first 5 rows of the DataFrame to check the data structure and content
```

```python
df.head()
```

```python
# Drop the 'description' and 'listed_in' columns as they are not needed
```

```python
df.drop(columns=['description', 'listed_in'], inplace=True)
```

```python
df.head()
```

```python
# Convert 'date_added' column to datetime format, coerce invalid dates to NaT (Not a Time)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.head()
```

```python
# Extract the year, month, and day from 'date_added' and create new columns for them
```

```python
df['year_added'] = df['date_added'].dt.year
```

```python
df['month_added'] = df['date_added'].dt.month
```

```python
df['day_added'] = df['date_added'].dt.day
```

```python
df.head()
```

```python
# Drop the 'date_added' column as it is no longer needed after extracting its components
```

```python
df.drop(columns=['date_added'], inplace=True)
```

```python
df.head()
```

```python
# Define a dictionary to map categorical ratings (e.g., TV-MA) to numerical age values
```

```python
rating_mapping = {'TV-MA': 18, 'TV-14': 14, 'TV-PG': 10, 'R': 18, 'PG-13': 13} 
```

```python
# Map the 'rating' column using the rating_mapping dictionary, fill missing values with 0, and convert to integers
```

```python
df['rating'] = df['rating'].map(rating_mapping).fillna(0).astype(int)
```

```python
df.head()
```

```python
# Extract the numeric part of the 'duration' column using a regular expression and convert it to float
```

```python
df['duration'] = df['duration'].str.extract(r'(\d+)').astype(float)
```

```python
df.head()
```

```python
# Define a list of columns to be converted to uppercase
```

```python
columns_to_convert = ['show_id', 'type', 'title', 'director', 'cast', 'country']
```

```python
# Convert the specified columns to strings and apply uppercase transformation
```

```python
df[columns_to_convert] = df[columns_to_convert].astype(str).apply(lambda x: x.str.upper())
```

```python
df.head()
```

```python
# Identify all numeric columns (float64 and int64 types) for missing value handling
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Fill missing values in numeric columns with the column's mean value
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df.head()
```

```python
# Identify all categorical columns (object type) for missing value handling
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# Fill missing values in categorical columns with the most frequent value (mode)
```

```python
df.update(df[categorical_cols].apply(lambda col: col.fillna(col.mode()[0])))
```

```python
df.head()
```

```python
# Print the entire DataFrame after cleaning and transformation
```

```python
print(df)
```

```python
# Repeat the process to ensure all numeric columns have their missing values filled
```

```python
numeric_cols = df.select_dtypes(include=['number']).columns
```

```python
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
# Ensure all categorical columns have their missing values filled with the mode
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
```

```python
# Print the first 5 rows of the cleaned DataFrame to verify the final state
```

```python
df.head()
```

```python
# Save the cleaned DataFrame to a new CSV file, without including the index
```

```python
df.to_csv("netflix_data1.csv", index=False)
```
